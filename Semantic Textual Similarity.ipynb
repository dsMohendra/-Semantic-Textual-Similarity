```python
# i want to splitting the data into standrade way , i dont want to loose the order
x_row =int(np.round(df_text_data[['text1','text2']].shape[0]*(0.80))) # this is dependent column 
Xtrain = df_text_data[['text1','text2']][: x_row]
Xtest = df_text_data[['text1','text2']][x_row :]

y_row = int(np.round(df_text_data['textual similarity'].shape[0]*0.80))
ytrain=df_text_data['textual similarity'][:y_row]
ytest = df_text_data['textual similarity'][y_row:]

print(Xtrain.shape)
print(Xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
### Tokenizer (Tokenize the word into number ) 
# creat tokenizer object
obj_token = Tokenizer(num_words=num_corpus)
#Tokenizer mainly used for words not for sentence

obj_token.fit_on_texts(Xtrain)
t_Xtrain=obj_token.texts_to_sequences(Xtrain)
t_Xtest=obj_token.texts_to_sequences(Xtest)
print('__After tokenize the Xtrain,Xtest__')
# print(t_Xtrain[np.random.randint(1,400)])
print('PAD_SEQUENCE __it will standize the length of words into same ')# like image processing
t_Xtrain = pad_sequences(sequences=t_Xtrain,
    maxlen=128,
    dtype='int32',
    padding='post',
    truncating='post')
t_Xtest = pad_sequences(sequences=t_Xtest,
    maxlen=128,
    dtype='int32',
    padding='post',
    truncating='post')
t_Xtest
```

### Creating Sequential Model by usening LSTM

```python
rnn_model = Sequential()

def rnn_lstm_model():
    rnn_model.add(Embedding(input_dim= num_corpus,
        output_dim=100,
        embeddings_initializer='uniform',
        embeddings_regularizer=None,
        activity_regularizer=None,
        embeddings_constraint=None,
        mask_zero=False,
        input_length=128))
    rnn_model.add(LSTM(units=100,
        activation='tanh',
        recurrent_activation='sigmoid',
        use_bias=True,
        kernel_initializer='glorot_uniform',
        recurrent_initializer='orthogonal',
        dropout=0.1))
    #output layer
    rnn_model.add(Dense(units=1,
        activation='sigmoid'))
    rnn_model.summary()
    
    rnn_lstm_model_compile()

def rnn_lstm_model_compile():
    print('Compiler Transforming Source Code to Machine Code')
    
    rnn_model.compile(optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'])   
rnn_lstm_model()
```

```python
def rnn_lstm_model_fit_evaluete():
    print('IT train to the model then it do evaluation')
    rnn_model.fit(Xtrain,ytrain,epochs=10)
    print()
    print('Evaluation of model loss  , accuracy ')
    rnn_model.evaluate(Xtest,ytest)
    
# rnn_lstm_model_fit_evaluete()
```
